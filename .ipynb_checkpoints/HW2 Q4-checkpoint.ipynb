```julia
######### Question 4 #########
using DataFrames, CSV, LinearAlgebra
using StatsModels

Trees = CSV.read("cherry-trees.csv", DataFrame)

fo = @formula(Volume ~ 1 + Girth)
A = modelmatrix(fo.rhs, Trees)      # the predictors
b = vec(modelmatrix(fo.lhs, Trees)) # the response
    

function huber(u; M=1)
    if norm(u) <= M
        h = u^2
    else
        h = M*(2*norm(u) - M)
    end
    return h
end
    
function h(x; M=1)
    uvec = A*x - b
    sum = 0
    for m in range(start = 1, stop = length(uvec), step = 1)
        u = uvec[m]
        sum += huber(u, M=M)
    end
    return sum
end
    
function ∇h(x; M=1)
    uvec = A*x - b
    ∇hub = [Float64(0), Float64(0)]
    for m in range(start = 1, stop = length(uvec), step = 1)
        u = uvec[m]
        if norm(u) <= M
            ∇hub[1] += 2u*A[m, 1]
            ∇hub[2] += 2u*A[m, 2]
        else
            ∇hub[1] += 2M*A[m, 1]*sign(u)
            ∇hub[2] += 2*M*A[m, 2]*sign(u)
        end
        
    end
    return ∇hub
end

function line_search(f, ∇f, x, p; strong=true, c1=1e-4, c2=0.9, rho=0.75)
    α = 1    # initial step length = 1 required for newton-like methods
    if strong == true
        while f(x + α*p) > f(x) + c1*α*∇f(x)'*p || abs(dot(∇f(x + α*p),p)) > c2*abs(dot(∇f(x),p))
            α = rho * α
        end
    else
        while f(x + α*p) > f(x) + c1*α*transpose(∇f(x))*p || transpose(∇f(x + α*p))*p < c2*transpose(∇f(x))*p
            α = rho * α
        end
    end
    return α
end

# steepest descent using a line search method to determine step size
function steepest_descent(f, ∇f, x; ϵ=1e-1, k=10)
    i = 1
    while norm(∇f(x)) > ϵ
        # compute the direction
        B = -I       # steepest descent
        p = B * ∇f(x)
        
        # determine the step size
        α = line_search(f, ∇f, x, p)

        x = x + α*p  # the new iterate
        i % k == 0 && println("iteration ", i, ": step size = ", α, ", x = ", x)
        i += 1
    end
    return x
end

x0 = [-29, 4.4]
hl = steepest_descent(h, ∇h, x0, ϵ=1e-1, k=100)
```

```julia
using Plots
scatter(A[:,2], b, label="", xlabel="Girth", ylabel="Volume", fmt = :png)
f(x) = hl[1] + hl[2]*x
plot!(f, label = "huber loss", legend = :bottomright)
```
